# 数据切分

```python
X = model_data[analysis_cols]
y = model_data['waybill_label'] 
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=2023)        
print("x_train.shape: {}".format(x_train.shape))                                                        
print("x_test.shape: {}".format(x_test.shape))  
```

# xgboost模型

```python
import pickle
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, make_scorer, r2_score
from sklearn.model_selection import KFold, train_test_split, GridSearchCV 
from scipy.sparse import csr_matrix, hstack   
```

```python
grid_params = {'learning_rate': [0.05, 0.1, 0.3], 'max_depth': [5, 6, 8], 'min_child_weight': [5, 7, 9], 'gamma': [0, 0.1], 'reg_alpha': [0.01, 0.05]}
main_params = {
    'seed': 0,
    'n_estimators': 10,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
}
est = xgb.XGBClassifier(**main_params)   
```

- 网格搜索交叉验证

```python
gridsearch_cv = GridSearchCV(est, param_grid=grid_params, cv=3, scoring='roc_auc', n_jobs=cpu_count() // 2 + 1) 
gridsearch_cv.fit(x_train, y_train)   
```

```python
evalute_result = gridsearch_cv.cv_results_
print('每轮迭代运行结果:{}'.format(evalute_result))
print('参数的最佳取值：{}'.format(gridsearch_cv.best_params_))
print('最佳模型得分: {}'.format(gridsearch_cv.best_score_))
print("best_estimator_: ")
print(gridsearch_cv.best_estimator_)           
```

- 最优模型

```python
# 模型路径
brand_name = "FakeReceipt"
model_path = "./model_save/" + brand_name + "_xgboost_regression.pkl"
# 获取最优参数后训练大模型
params_map = {**main_params, **gridsearch_cv.best_params_}
print("params_map: {}".format(params_map))
xgb_model_best = xgb.XGBClassifer(**params_map)
xgb_model_best.fit(x_train, y_train)       
joblib.dump(xgb_model_best, model_path)     
print("模型训练以及保存完成。")           
```

```python
fmap = {}   # 特征映射
for i in range(len(analysis_cols)):
    fmap['f'+str(i)] = analysis_cols[i]
print(fmap)   
```

```python
score = xgb_model_best.get_booster().get_score(importance_type='gain')    # 只包含非0特征
print(score)
score_tmp = [(k, v) for k, v in score.items()] 
score_df = pd.DataFrame(score_tmp, columns=['features', 'gain_value'])
score_df['features'] = score_df['features'].apply(lambda x: fmap[x])
score_df.sort_values(by='gain_value', ascending=False, inplace=True)
print(score_df)  
```

```python
def bar_plot(x_col, y_col, df, orient_type, whether_text_annotate=True):
    """
    单离散特征-直方图, 支持水平直方图、垂直直方图。
    :param x_col: str, x轴名称
    :param y_col: str, y轴名称
    :param df: DataFrame, 数据源
    :param orient_type: str, 'h' -> 水平直方图, 'v' -> 垂直直方图.
    水平直方图，x为类别型特征，垂直直方图，y为类别型特征
    :param whether_text_annotate: 是否标注文字，默认False
    :return:
    """
    fig = sbn.barplot(x_col, y_col, data=df, orient=orient_type)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title('{}-{} 直方图'.format(x_col, y_col))
    if whether_text_annotate:
        for p in fig.patches:
            if p.get_width() > 0:
                fig.annotate(
                    text=f"{p.get_width(): 0.3f}",  # 文字内容
                    xy=(p.get_width(), p.get_y() + p.get_height() / 2.),  # 文字的位置
                    xycoords='data', ha='center', va='center', fontsize=10., color='black',
                    xytext=(0, 0),  # 文字的偏移量
                    textcoords='offset points', clip_on=True)
        plt.tight_layout()
    plt.savefig('./model_save/plot_feature_importance.svg', dpi=500., bbox_inches='tight', format='svg')
    plt.show()   # show之前保存
```

```python
bar_plot(x_col='gain_value', y_col='features', df=score_df, orient_type='h')  
```
