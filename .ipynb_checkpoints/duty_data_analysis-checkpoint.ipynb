# 特征工程

```python
def feature_engineer(df, factor_cols, fillna_cols=None, whether_fillna=True):
    '''因子型特征编码'''
    from sklearn.preprocessing import LabelEncoder
    # 空值填充
    if whether_fillna and fillna_cols:   
        df[fillna_cols].fillna('否', inplace=True)
    # 数值化
    for col in factor_cols:
        tmp = df[col].astype(str).apply(lambda x: 0 if str(x) == '否' else 1)
        df[col] = tmp
        print(df[col].unique())
        #enc = preprocessing.LabelEncoder()
        #enc = enc.fit(['是', '否'])
        #df[col] = enc.transform(df[col])
    # 网点百票罚款
    df['100t_hand_amount'] = df['hand_amount'] * 100
    # 网点下车和签收时间的时间间隔
    df['down_sign_time'] = np.where(df['down_sign_time']<0, 0, df['down_sign_time'])
    # log网点近7天集中签收量 
    # df['log_sign_7focus_num'] = df['sign_7_focus_num'].map(np.log)
    df.drop(['hand_amount'], axis=1, inplace=True)
    return df 
```

```python
analysis_col_dict = {'down_sign_time': '下车与签收时间间隔',
'hand_num': '网点日均派件量',
# 'is_profit': '是否盈利',
# 'miss_sign_7avg_num': '网点近七天日均遗失补签量',
# 'hand_sign_num': '业务员当日签收票数',
# '751_7avg_rate': '业务员近7天入库入柜率',
# 'complaint_7avg_num': '业务员近7天日均投诉量',
}
```

```python
def plot_num_feature(df, analysis_col_dict, label_col):
    for k, v in analysis_col_dict:
        print('{}: {}'.format(k, v))
        plt.subplot(2, 1, 1)
        sbn.boxplot(data=df, x=label_col, y=k)
        plt.xlabel(v, fontdict={'weight':'normal', 'size': 10})
        plt.xlabel('虚假签收标签', fontdict={'weight':'normal', 'size': 10})
        plt.legend()
        plt.subplot(2, 1, 2)
        fake = df[df[label_col] == 1][k]
        non_fake = df[df[label_col] == 0][k]
        sbn.kdeplot(fake, shade=True, label='虚假签收', color='hotpink')
        sbn.kdeplot(non_fake, shade=True, label='非虚假签收', color='seagreen')
        plt.savefig("./feature_visualize/{}.png".format(v), dpi=1000)
        plt.show()    
```

```python
plot_num_feature(df_feat, analysis_col_dict, label_col='waybill_label')
```
